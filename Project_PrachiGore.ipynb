```python
################### 
import urllib.request, urllib.parse, urllib.error
from bs4 import BeautifulSoup
import ssl
import urllib.request
import requests
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.77 Safari/537.36'}

#creating empty lists

director_list=[]      #to store multiple director names
url_list = []         #to store all main page URL
list_of_movies=[]     #final list of dictionary structure of each movie
mov_list=[]           #list of movie links


for i in range(1,43):   #looping through main page urls
    url = 'https://www.metacritic.com/browse/movie/?releaseYearMin=1910&releaseYearMax=2023&page=' + str(i)
    url_list.append(url)    #creating a list of url of every page.
    

for i in url_list:      #iterating over the main page urllist
    html = requests.get(i, headers = headers)
    soup = BeautifulSoup(html.content, "html.parser")
    tags = soup.find_all('a', class_ = "c-finderProductCard_container g-color-gray80 u-grid") #using soup to get movie link
    for movie in tags:
        movie_page = 'https://www.metacritic.com' + movie['href']
        mov_list.append(movie_page)   #appending each movie link to the mov_list

n=0
for i in mov_list:   #iterating over the list of each movie links
    n = n +1 
    print(n)
    movie_page = requests.get(i, headers = headers) 
    movie_page_soup = BeautifulSoup(movie_page.content, "html.parser") 
    movie_name_list = movie_page_soup.find_all('div', class_ = "c-productHero_title g-inner-spacing-bottom-medium g-outer-spacing-top-medium")
    
       
    for i in movie_name_list:
        movie_name = i.get_text().strip() 
        #print('movie name:',movie_name)

        view_all = movie_page_soup.find_all('a', class_ = "c-globalHeader_viewMore g-color-gray50 g-text-xxsmall u-text-uppercase")
    
    for i in view_all:
        view_all_link = 'https://www.metacritic.com' + i['href'] #extract the url to the "view all" page
        #print(view_all_link)
    
    view_all_page = requests.get(view_all_link, headers = headers) #using bs4 to scrape the "view all" page
    view_all_soup = BeautifulSoup(view_all_page.content, "html.parser")

    credit_tables_list = view_all_soup.find_all('div', class_ = "c-productCredits g-outer-spacing-bottom-xlarge")

    for i in credit_tables_list:

        if "Directed By" in i.get_text().strip(): #Adding condition using key word Directed by 
            director_names = i.find_all('a')  #if the condition is true it will find all the director names
            director_list=[] #creating empty list for storing director names
            for j in director_names:
                director_list.append(j.get_text().strip())   
                director_name=tuple(director_list)  #converting list to tuple to handle multiple director names for single movie      
    
    casts = []    
    for i in credit_tables_list: # Iterating  over the credit table list to fetch cast the movie

        table_title = i.find('h3', class_ = "c-productCredits_groupName g-color-gray80 g-text-bold u-text-uppercase")
        #table_title will have only the cast names with heading cast not others ex- cast director table
        #print(table_title)
        if table_title.get_text().strip()=="Cast":  #condition to check the cast table using 'Cast' keyword
            cast_names = i.find_all('a')  #finding all cast names from cast table and iterating over it in below for loop
            
            for j in cast_names:
                cast_name = j.get_text().strip() #stripping the unwanted details from cast names and adding them into the casts list as below
                casts.append(cast_name)
    
    dict_1000 = {} #creating an empty dictionary to store the extracted data
    
    
    for i in mov_list: #getting all the 3 variables from each movie
        
        sub_dict = {}              #creating a subdictionary to hold director name as key and cast as value 
        sub_dict[director_name] = casts       #assigning keys-director name and values-casts list  
        dict_1000[movie_name] = sub_dict     #setting movie name as a key and sub dictionary as one value

        #if dict_1000 not in list_of_movies: #added if condition to see if the dict does not exists already
            
    list_of_movies.append(dict_1000) #appending the dictionary into a list 
#print(len(list_of_movies),list_of_movies)      #print statementstotext.
```
